```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from yellowbrick.regressor import ResidualsPlot

df = pd.read_csv('CO2_Emissions_Canada.csv')
sns.set(rc={'figure.figsize':(20,13)})
df = df.dropna()

plt.figure(figsize = (13,9))
sns.boxplot(x="Make", y="CO2 Emissions(g/km)", data=df)
plt.xticks(rotation = 90)
plt.show()

sns.jointplot(
    data=df,
    x="Fuel Consumption Comb (mpg)", y="CO2 Emissions(g/km)", hue="Fuel Type"
)
plt.show()

cor = df.corr()
heatmap = sns.heatmap(cor, annot=True, cmap='PiYG')
heatmap.set_title('Heatmap', fontdict={'fontsize':20}, pad=10);

num_features = [feature for feature in df.columns if df[feature].dtype != '0']
discrete_features = [feature for feature in num_features if len(df[feature].unique()) < 25]

for feature in discrete_features:
  data = df.copy()
  print(data.groupby(feature)['CO2 Emissions(g/km)'].sum() / data['CO2 Emissions(g/km)'].sum())
  data.groupby(feature)['CO2 Emissions(g/km)'].mean().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Emission')
  plt.show()
```
