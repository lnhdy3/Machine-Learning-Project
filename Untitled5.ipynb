```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)
from sklearn import preprocessing
from datetime import datetime, timedelta
import random
from sklearn.preprocessing import OneHotEncoder
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

from sklearn.tree import DecisionTreeClassifier
import matplotlib.image as pltimg

from sklearn.model_selection import train_test_split
from yellowbrick.regressor import ResidualsPlot
from sklearn.pipeline import Pipeline

df = pd.read_csv('CO2_Emissions_Canada.csv') 
df2 = df.copy()


df.isnull().sum()

#Q1
#Examining a regression problem so plotting target value
#in data set 

sns.displot(data = df, x = 'CO2 Emissions(g/km)')
plt.show()
print(df.describe())

#Q2
#Look for null values, no null values found 

df.isnull().any()
df['CO2 Emissions(g/km)'].isnull().values.any()
#df.info()
plt.show()

#sns.displot(df, x="CO2 Emissions(g/km)", hue="Fuel Consumption Comb (mpg)")

#Q3 drop columns not useful for model training

df = df.drop(columns = ['Engine Size(L)', 'Model', 'Transmission'])


#Q4 determine which columns should be label-encoded, and which
#should be one-hot-encoded
#label-encoding
le = preprocessing.LabelEncoder()
df['Make'] = le.fit_transform(df['Make'])
df['Vehicle Class'] = le.fit_transform(df['Vehicle Class'])
df['Fuel Type'] = le.fit_transform(df['Fuel Type'])


#one-hot-encoding
fuel_types = ('Z','X','N','E', 'D')
fuel_df = pd.DataFrame(fuel_types, columns=['Fuel_Types'])
dum_df = pd.get_dummies(fuel_df, columns=["Fuel_Types"], prefix=["Type_is"] )
fuel_df = fuel_df.join(dum_df)
display(fuel_df)



#Q5 Create at least three agression columns, I chose to do mean and
#standard deviation
df['Mean CO2 Emissions by Make'] = df.groupby('Make')['CO2 Emissions(g/km)'].transform('mean')
df['Mean CO2 Emissions by Vehicle Class'] = df.groupby('Vehicle Class')['CO2 Emissions(g/km)'].transform('mean')
df['Standard Deviation of CO2 Emissions by Fuel Consumption'] = df.groupby('Fuel Consumption Comb (mpg)')['CO2 Emissions(g/km)'].transform('std')
df['Standard Deviation of CO2 Emissions by Fuel Consumption'] = df['Standard Deviation of CO2 Emissions by Fuel Consumption'].fillna(value = 0)

#Q6
#Q7 Split Data into features (x_data) and labels (y_data)
y_data = df['CO2 Emissions(g/km)']
x_data = df[['Make','Vehicle Class', 'Cylinders', 'Fuel Type',
             'Fuel Consumption City (L/100 km)', 'Fuel Consumption Hwy (L/100 km)', 
             'Fuel Consumption Comb (L/100 km)', 'Fuel Consumption Comb (mpg)']]

y_columns = ['CO2 Emissions']
x_columns = x_data.columns.tolist()

#Q8 scale data
#scaler = preprocessing.StandardScaler()
scaler = StandardScaler()
x_data = scaler.fit_transform(x_data)


#Q9
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.1, random_state = 1)
#model
lin_reg = linear_model.LinearRegression()
lin_reg = lin_reg.fit(x_train, y_train)
lin_reg.score(x_test, y_test)

fig = ResidualsPlot(lin_reg)
fig.fit(x_train, y_train)
fig.score(x_test, y_test)
fig.show()

sgd_reg = linear_model.SGDRegressor()
sgd_reg = sgd_reg.fit(x_train, y_train)
sgd_reg.score(x_train, y_train)

fig = ResidualsPlot(sgd_reg)
fig.fit(x_train, y_train)
fig.score(x_test, y_test)
fig.show()



#Part 2



#1. if you are aiming for classification, import the supervised learning models, linear
#regression and decision tree classifier from sklearn. If regression, import linear regression
#and logistic regression.
#2. Fit / train both models on the training dataset / labels
```
